```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Image URL to search for
image_url = "https://assets.philosophie.ch/2023-08-10-erni.jpeg"

# Set up Selenium WebDriver
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")  
service = Service("/home/alebg/Descargas/chromedriver-linux64/chromedriver")
driver = webdriver.Chrome(service=service, options=options)

wait = WebDriverWait(driver, 10)  # Wait up to 10 seconds

try:

    # Step 1: Open Google Images
    driver.get("https://www.google.com/imghp")
    time.sleep(2)

    popup = WebDriverWait(driver, 5).until(
        EC.presence_of_element_located((By.XPATH, "//div[contains(@class, 'popup')]"))
    )
    popup.click()


    time.sleep(2)  # Wait a bit for the popup to disappear
    # Step 2: Click on "Search by Image" (Camera Icon)
    search_by_image_button = driver.find_element(By.CLASS_NAME, "Gdd5U")
    search_by_image_button.click()
    time.sleep(2)

    # Step 3: Switch to "Paste Image URL" tab
    paste_url_tab = driver.find_element(By.XPATH, "//a[contains(text(),'Paste image link')]")
    paste_url_tab.click()
    time.sleep(2)

    # Step 4: Paste the Image URL and Search
    url_input = driver.find_element(By.XPATH, "//input[@type='text']")
    url_input.send_keys(image_url)
    url_input.send_keys(Keys.RETURN)
    time.sleep(5)  # Wait for results to load

    # Step 5: Click "Tools" and Apply "Usage Rights" Filter
    tools_button = driver.find_element(By.XPATH, "//div[text()='Tools']")
    tools_button.click()
    time.sleep(2)

    usage_rights_menu = driver.find_element(By.XPATH, "//div[text()='Usage rights']")
    usage_rights_menu.click()
    time.sleep(2)

    # Select "All" to display copyright info in results
    all_usage_rights = driver.find_element(By.XPATH, "//span[text()='All']")
    all_usage_rights.click()
    time.sleep(5)  

    # Step 6: Extract Search Results & Check for Copyright Terms
    results = driver.find_elements(By.CSS_SELECTOR, "div.MjjYud a")

    print("\nüîç Possible Copyrighted Images:")
    copyright_terms = ["¬©", "All Rights Reserved", "Getty", "Shutterstock", "Adobe Stock", "AP Photo"]

    for i, result in enumerate(results[:5]):  # Limit to top 5 results
        title = result.text
        link = result.get_attribute("href")

        # Flag if the title contains known copyright indicators
        if any(term in title for term in copyright_terms):
            print(f"‚ö†Ô∏è Copyrighted: {title} - {link}")
        else:
            print(f"‚úÖ Not Flagged: {title} - {link}")

finally:
    driver.quit()
```
